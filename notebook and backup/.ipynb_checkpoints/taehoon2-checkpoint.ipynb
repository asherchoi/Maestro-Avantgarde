```python
def gram_matrix(filters):
    c_filters = K.batch_flatten(K.permute_dimensions(K.squeeze(filters, axis=0), pattern=(2, 0, 1)))
    return K.dot(c_filters, K.transpose(c_filters))
```

```python
A = [gram_matrix(o) for o in style_features]
```

```python
noise_features = model.predict(noise_pre)
```

```python
G = [gram_matrix(o) for o in noise_features]
```

```python
style_layers = layers[:]
style_loss = 0.0
style_layer_weight = 1.0 / float(len(style_layers))
```

```python
for i in range(len(style_layers)):
    print(style_features[i].shape)
```

```python
N = style_features[0].shape[3]
M = style_features[0].shape[1] * style_features[i].shape[2] 
print('1_layer filter number:', N)
print('1_layer feature map size:', M)
```

```python
for i, style_layer in enumerate(style_layers):
    N = style_features[i].shape[3]
    M = style_features[i].shape[1] * style_features[i].shape[2] 
    style_loss += (style_layer_weight *  (1.0 / (4.0 * (N ** 2.0) * (M ** 2.0))) * K.sum(K.square(G[i] - A[i])))
    
beta = 1
total_loss = style_loss * beta
print(total_loss)
```

```python
content_features = model.predict(content_pre)[]
```

```python
content_loss = 0.5 * K.sum(K.square(noise_features - content_features))
```

```python
grads = K.gradients(total_loss, noise_pre)
```

```python
from keras import losses
```

```python
model.compile(loss=losses.mean_squared_error, optimizer='sgd', metrics=total_loss)
```
